```python
import os
import zipfile

def extract_all_zip_files_in_folder(folder_path, destination_folder):
    # Get all files in the folder
    files = os.listdir(folder_path)

    # Filter out only ZIP files
    zip_files = [f for f in files if f.endswith('.zip')]

    # Extract each ZIP file
    for zip_file in zip_files:
        zip_file_path = os.path.join(folder_path, zip_file)
        with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
            zip_ref.extractall(destination_folder)
```

```python
import pandas as pd
import os

file_path = "/content/drive/My Drive/IE434 Deep Dive 11 project/working_df.pkl"
if os.path.exists(file_path):

    pass  # or do whatever you want when the file exists
else:
  folder_path = '/content/drive/MyDrive/IE434 Deep Dive 11 project/Deep Dive 2 (Milestone 1)/data zip'
  destination_folder = '/content/drive/MyDrive/IE434 Deep Dive 11 project/Deep Dive 2 (Milestone 1)/data_extracted'
  extract_all_zip_files_in_folder(folder_path, destination_folder)

  dir_path = '//content/drive/MyDrive/IE434 Deep Dive 11 project/Deep Dive 2 (Milestone 1)/data_extracted'

  # Get all files from the directory with a .csv extension
  all_files = [os.path.join(dir_path, file) for file in os.listdir(dir_path) if file.endswith('.csv')]

  # Initialize an empty list to store individual dataframes
  dfs = []

  # Loop through all files and read them into dataframes
  for file in all_files:
      df = pd.read_csv(file)
      dfs.append(df)

  # Concatenate all dataframes to create a master dataframe
  master_df = pd.concat(dfs, ignore_index=True)

  # Display the master dataframe
  print(master_df.head())
  master_df.to_pickle("/content/drive/MyDrive/IE434 Deep Dive 11 project/Deep Dive 2 (Milestone 1)/working_df.pkl")
```
